# Informal organisation - social network analyses

In this research I aimed to answer the question how a firm’s internal structure hindersinformation flow. To measure the efficiency of information flow I used communication quality which was calculated by averaging others feedbacks. The formal and the informal network were created by questionnaire focusing on information flow. From these variables I have built a graph, where the nodes were the employees and edges showed how information flows between them. To signal strength of connection I  calculated a certain weight for each edge based on one’s nominating affinity. The nodes attributes included their corporal position, function, department and region. The research took place in a Hungarian logistic firm with 6000 thousand employees, response rate for the questionnaire was around 50% equally distributed within hierarchy. Lastly, I measured one’s social skills
and corporate skills based on the questionnaire, by summing up nominations on the relevant parts. 

One of the methodological challenges was the skewedness of the data. The distribution of the in and out degrees were expected to be skewed based on the social networks scale-free distribution, but many other variables (communication quality, social and corporate skills, network attributes) were far from normal distribution. As I intended to use OLS regression to answer how the number of degrees a person has affects their communication skill, I had to transform several variables to reach the underlying assumptions of best linear unbiased estimator. I have chosen Box – Cox transformation which changed the variables distribution to a more normal shape.

```python
%load_ext autoreload
%autoreload 2
import networkx as nx
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt 
from sklearn import preprocessing
import matplotlib.patches as mpatches
import nx_functions as f
```

```python
def prob_plot(df, before, after):
    # Communication quality
    fig = plt.figure()
    ax1 = fig.add_subplot(211)
    x = df[before].values
    # Before transformation
    prob = stats.probplot(x, dist=stats.norm, plot=ax1)
    ax1.get_lines()[0].set_markerfacecolor('tab:blue')
    ax1.get_lines()[0].set_markeredgecolor('tab:blue')
    ax1.set_title(str(before) + ' against normal distribution')

    ax2 = fig.add_subplot(212)
    xt = df[after].values
    prob = stats.probplot(xt, dist=stats.norm, plot=ax2)
    ax2.get_lines()[0].set_markerfacecolor('tab:blue')
    ax2.get_lines()[0].set_markeredgecolor('tab:blue')
    ax2.set_title(str(before) + ' after Box-Cox transformation')
    plt.tight_layout()
    plt.show()
```

```python
def my_hist(df, col, lower=0.05, higher=0.95):
    low_ = df[col].quantile(lower)
    high_ = df[col].quantile(higher)
    plt.hist(df[col], color='tab:blue')
    plt.hist(df[col], histtype='step', color='tab:red', cumulative=True)
    plt.axvline(low_, linestyle=':', color='tab:green')
    plt.axvline(high_, linestyle=':', color='tab:green')
    plt.title('Distribution of ' + str(col))
    plt.show()
    return plt
```

```python
def distribution_of_answers(df, col, dir_, rot_):
    if dir_ == 'source':
        _dir = 'to'
    else:
        _dir = 'source'
    df_all = df[[dir_, col]].groupby([col]).count()
    df_all.columns = ['n']
    dist = df[[dir_, 'question', col, _dir]].groupby(
        [col, dir_, 'question']).count()
    df_emp = df[[dir_, col]].groupby(
        [col, dir_]).count().reset_index().groupby([col]).count()
    df_emp.columns = ['employees']
    df_final = df_all.join(df_emp)
    df_final['ratio'] = df_final['n']/df_final['employees']/17

    # boxplot

    dist = dist.reset_index()
    dist.drop('question', axis=1, inplace=True)
    dist.fillna(0, inplace=True)
    dist.boxplot(by=col, rot=rot_)
    plt.tight_layout()
    plt.show()
    return df_final
```

```python
def scatter_it(df, x_, y_, xmin, xmax, ymin, ymax, size, trend=False):
    x = df[x_]
    y = df[y_]
    color_dict = {'E': 'blue', 'TL': 'red', 'MM': 'orange', 'TM': 'green'}
    plt.scatter(x, y, color=[color_dict[i] for i in df['hierarchy']], s=size)
    plt.axis([xmin, xmax, ymin, ymax])
    markers = [plt.Line2D([0, 0], [0, 0], color=color, marker='o',
                          linestyle='') for color in color_dict.values()]
    plt.legend(markers, color_dict.keys(), numpoints=1)

    if trend:

        # overall trend
        z = np.polyfit(x, y, 1)
        p = np.poly1d(z)
        plt.plot(x, p(x), color='black', linestyle='dashed')
        plt.text(xmax*0.9, ymax*0.3, "y=%s+%s" %
                 (np.round(z[0], 2), np.round(z[1], 2)), ha='right')

        for i, key in enumerate(color_dict.keys()):
            try:
                x = df[df.hierarchy == key][x_]
                y = df[df.hierarchy == key][y_]
                z = np.polyfit(x, y, 1)
                p = np.poly1d(z)
                plt.plot(x, p(x), color=color_dict[key], linestyle=':')
                plt.text(xmax*0.9, (ymax*0.2)-i*ymax*0.05, "y=%s+%s" %
                         (np.round(z[0], 2), np.round(z[1], 2)), ha='right', color=color_dict[key])
            except:
                print(key)
    plt.xlabel(x_)
    plt.ylabel(y_)
    plt.title('Scatter plot of :' + str(x_) + ','+str(y_))
    plt.show()
```

# Reading in original edges and questions, filter by information questions create weights to it. 

```python
# Reading edges in
edges = pd.read_csv('110_answers.csv')
edges = edges[edges.answer.notnull()]
edges = edges[['from_uid', 'answer', 'question_dimension', 'Communication']]
edges.columns = ['source', 'to', 'question', 'score']
edges.set_index('question', inplace=True)
# Categorizing them
categories = pd.read_csv('influence_questions.csv')
categories.set_index('question', inplace=True)

# Join
edges = edges.join(categories)
edges.reset_index(inplace=True)
edges.set_index('source', inplace=True)
# edges=edges[edges.segment_a!='attitude']
edges = edges[['to', 'question', 'segment_b', 'score']]


# informal_edges
informal_edges = edges[edges.segment_b == 'information']
informal_edges = informal_edges.groupby(
    ['source', 'to']).agg({'segment_b': 'count'})
informal_edges.reset_index(inplace=True)
informal_edges.columns = ['src', 'dst', 'weight']
informal_edges.to_csv('informal_edges_raw.csv')

# formal edges
formal_edges = edges[edges.segment_b == 'formal'][['to', 'score']]
formal_edges.to_csv('formal_edges_raw.csv')
```

# Generating betcet and closecent

```python
G=nx.from_pandas_edgelist(informal_edges,'dst','src', edge_attr=True, create_using=nx.DiGraph())
```

```python
print (nx.info(G))
```

```python
in_degree_df = pd.DataFrame.from_dict(dict(G.in_degree()), orient='index')
in_degree_df.columns = ['in_degree']
out_degree_df = pd.DataFrame.from_dict(dict(G.out_degree()), orient='index')
out_degree_df.columns = ['out_degree']
degree_df = in_degree_df.join(out_degree_df, how='outer')
degree_df['sum_degree'] = degree_df.in_degree + degree_df.out_degree
```

```python
degree_df
```

```python
betcent_1=nx.betweenness_centrality(G,weight='weights')
bc_1=pd.DataFrame.from_dict(betcent_1, orient='index')/betcent_1[max(betcent_1, key=betcent_1.get)]
bc_1.columns=['bc_1']
```

```python
clocent=nx.closeness_centrality(G)
clocent=pd.DataFrame.from_dict(clocent, orient='index')/clocent[max(clocent, key=clocent.get)]
clocent.columns=['clocent']
```
